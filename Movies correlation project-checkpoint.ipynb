# Cleaning up our dataset


### Update numeric data types for cleaner presentation

We find that the 'budget' column above has the dtype float64, which appends a '.00' to the end of each value in the column. We can remove the unnecessary decimal places with the Pandas .astype 

### Insert new column with corrected release years according to "year" column

The year in the 'released' column doesn't match the year in the 'year' column, this may or may not affect our analysis, but we can correct this by converting the data in 'released' from an object to a string, slicing the string at the first four digits (the year,) and assigning this to a new column that indicates the correct year of release. ***This step is not necessary if year is not included in your analysis.***

### Order rows by gross revenue

Using the method .sort_values, we can organize our data ordered by the 'gross' column, ordering from highest to lowest gross revenue.

### Drop duplicate rows

The .drop_duplicates method will remove duplicate entries. Compare the number of rows in the output to the number of rows in the original dataset to determine how many duplicate entries were deleted. ***In this case we find that no rows were deleted***

```python
# Update numeric data types for cleaner presentation
df['budget'] = df['budget'].astype('int64')
```

```python
# Insert new column with corrected release years according to "year" column
df['year_correct'] = df['released'].astype(str).str[:4]
```

```python
# Order rows by gross revenue
df = df.sort_values(by='gross', inplace = False, ascending = False)

df
```

```python
# Drop duplicate rows

df.drop_duplicates().sort_values(by = 'name', inplace = False, ascending = False)
```

# Data exploration to find correlations between budget, company, and revenue using scatter plots

Now we introduce some exploratory visualization. Using PyPlot from the MatPlotLib library, we can plot the relationship between 'budget' and 'gross' columns, and make an inferrence about the correlation between the two values


```python
plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs. Gross Revenue')
plt.xlabel('Budget')
plt.ylabel('Gross Revenue')
plt.show()

df.head()
```

### Using Seaborn to plot budget v gross regression

In addition to the scatterplot above, we can further observe the relationship between the two columns by including a regression line within the scatter plot with the Seaborn method .regplot.

```python


sns.regplot(x='budget', y='gross', data = df, scatter_kws = {'color' : 'red'}, line_kws = {'color' : 'blue'})
```

### Pandas also has built in methods to describe correlations 

We can produce a matrix of column correlations with the .corr method, and subsequently apply that matrix to a PyPlot heatmap graph for ease of identifying weaker and stronger correlations.

***Below the .corr method 'pearson' is called, more information about pandas.DataFrame.corr methods can be found [here](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.corr.html)***

```python
df.corr(method = 'pearson')
```

```python
correlation_matrix = df.corr(method = 'pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title('Correlations of Movies DataFrame')
plt.xlabel('Columns')
plt.ylabel('Columns')

plt.show()
```

# Further exploration of the dataset

The above only examines correlations between numeric datatypes. If we would like to explore correlations involving columns that describe non-numeric information, such as 'genre' or 'company' we can assign unique numeric values to strings or objects to then compare with our already numeric values. 

### Assigning numeric values to non-numeric dtypes

```python
df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized.head(20)
```

### Displaying correlations across entire dataset, including originally non-numeric dtypes

```python
correlation_matrix = df_numerized.corr(method = 'pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title('Correlations of Movies DataFrame')
plt.xlabel('Columns')
plt.ylabel('Columns')

plt.show()
```

### Presenting correlations as a table

As an alternative to the correlation matrix. The Pandas method .unstack will present a table of correlations with each column and its respective correlations to every other column. From here, we can sort from low to high, and identify strong positive correlations as higher than a correlation of 0.5 or higher.

```python
correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]
high_corr
```
